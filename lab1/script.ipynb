```python
# imports
import numpy as np
import matplotlib.pyplot as plt
```

```python

func = lambda x : x**2 + 5
set_1 = np.linspace(-1, 1)
set_2 = np.linspace(-6, 6)
set_3 = np.linspace(0, 5)
```

```python
f, (axis_1, axis_2, axis_3) = plt.subplots(3, 1, figsize=(10, 10))
axis_1.plot(set_1, func(set_1)), axis_1.set_xlim(left=-1, right=1), axis_1.set_title("f(x), x [-1,1]")
axis_2.plot(set_2, func(set_2)), axis_2.set_xlim(left=-6, right=6), axis_2.set_title("f(x), x [-6,6]")
axis_3.plot(set_3, func(set_3)), axis_3.set_xlim(left=0, right=5), axis_3.set_title("f(x), x [0,5]")
```
