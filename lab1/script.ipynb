## Zadanie 3
Matplotlib - pyplot, rysowanie wykresów z etykietami osi, tytułem i legendą.

```python
# imports
import numpy as np
import matplotlib.pyplot as plt
```

```python

func = lambda x : x**2 + 5
set_1 = np.linspace(-1, 1)
set_2 = np.linspace(-6, 6)
set_3 = np.linspace(0, 5)
```

```python
f, (axis_1, axis_2, axis_3) = plt.subplots(3, 1, figsize=(10, 20))
axis_1.plot(set_1, func(set_1), label='f(x) = x**2 + 5'), axis_1.set_xlim(left=-1, right=1), axis_1.set_title("f(x), x [-1,1]"), axis_1.set_xlabel('x'), axis_1.set_ylabel('f(x)'), axis_1.legend()
axis_2.plot(set_2, func(set_2),label='f(x) = x**2 + 5'), axis_2.set_xlim(left=-6, right=6), axis_2.set_title("f(x), x [-6,6]"), axis_2.set_xlabel('x'), axis_2.set_ylabel('f(x)'), axis_2.legend()
_ = axis_3.plot(set_3, func(set_3), label='f(x) = x**2 + 5'), axis_3.set_xlim(left=0, right=5), axis_3.set_title("f(x), x [0,5]"), axis_3.set_xlabel('x'), axis_3.set_ylabel('f(x)'), axis_3.legend()
# _ suppresses output
```

1. plot() - rysowanie
2. set_xlim() - ograniczenia na osi x
3. set_xlabel() i set_ylabel() - etykieta osi x i y
4. legend() - automatycznie generowana legenda

## Zadanie 4
Wstęp do Pandas - tworzenie Dataframe

```python
import pandas as pd
# using data classes
from dataclasses import make_dataclass
import enum

class Sex(enum.Enum):
    Male = 1,
    Female = 2
    
    def __str__(self):
        return self.name

Person = make_dataclass("Person", [
    ('name', str),
    ('surname', str),
    ('age', int),
    ('sex', Sex)
])

df = pd.DataFrame([
    Person('Sushil', 'Oz', 25, Sex.Male),
    Person('Nydia', 'Asami', 52, Sex.Female),
    Person('Yama', 'Ramazan', 15, Sex.Male),
    Person('Olexiy', 'Engilram', 18, Sex.Male),
    Person('Karmen', 'Genitus', 99, Sex.Female)
])

# you can display either by typing the name of the variable containing the DataFrame,
# or by printing, but vs code formats the non-print output better
df
```
